```python
#import packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
#Read in Data file as .csv, create filepath as df, and display
filepath='Influx_Underway_Darwin_TN413_2023v1_0_.csv'
picoplank_df = pd.read_csv(filepath, delimiter= ',')
print(picoplank_df.columns.values)
display(picoplank_df)
#Plot prochloro. abund. on y axis
fig,ax1 = plt.subplots(figsize=(20,6))
pro_abund= np.genfromtxt(filepath,dtype=float,usecols=5, delimiter=',')
lat= np.genfromtxt(filepath,dtype=float,usecols=1, delimiter=',')
ax1.plot(lat,pro_abund, c='g', linestyle= '-', linewidth=3, marker='.')
ax1.grid()
ax1.set_title('Prochlorococcus Abundance and Nitrate Concentration for TN413 Cruise', fontsize=20)
ax1.set_xlabel('Latitude', fontsize=16)
ax1.set_ylabel('Prochlorococcus Cell Abundance (cells/ul)', fontsize=16)
ax1.legend(['Prochlorococcus'],loc='upper left',fontsize=14)
#add twin x axes for latitude
ax2 = ax1.twinx()
#bring in nitrate data and plot it on second y axis
NO3_darwin_clim= np.genfromtxt(filepath,dtype=float,usecols=7, delimiter=',')
ax2.plot(lat,NO3_darwin_clim, c='k', linestyle= 'dashed', linewidth=3, marker='.')
ax2.grid()
ax2.set_ylabel('Nitrate Concentration (mmol/L)', fontsize=16)
ax2.legend(['Nitrate'],fontsize=14)
#save figure as a .png file
plt.savefig('TN413 pro')
plt.show()
#Source to add legend font size: https://matplotlib.org/stable/api/legend_api.html
#Source to add twin ax https://matplotlib.org/stable/gallery/subplots_axes_and_figures/two_scales.html
```

```python
#Create Transect Map Figure using Cartopy
#importing packages
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
#Make map using .PlateCarree, center map at 180 degrees lon. Add caostlines,land,and ocean.
fig = plt.figure(figsize=(20,10))
ax=plt.axes(projection=ccrs.PlateCarree(central_longitude=180))
ax.coastlines(resolution='110m',color='k')
g1 = ax.gridlines(crs=ccrs.PlateCarree(),draw_labels=True,linewidth=2,color='gray',alpha=0.5,linestyle=':')
g1.xlabels_top=False
g1.ylabels_right=False
g1.xformatter= LONGITUDE_FORMATTER
g1.yformatter= LONGITUDE_FORMATTER
ax.add_feature(cfeature.LAND,color='beige')
ax.add_feature(cfeature.OCEAN,color='lightblue')
#Plot Nitrate data by lat and lon
NO3_data= picoplank_df['NO3_darwin_clim']
lat = picoplank_df['lat']
lon = picoplank_df['lon']
Nutri= ax.scatter(lon,lat,c= NO3_data,transform=ccrs.PlateCarree(),cmap='plasma')
#Add colorbar with label, use set_extent to size map window the transect location of TN413
conc= plt.colorbar(Nutri)
conc.set_label('Nitrate Concentration (mmol/L)')
extent = [115,-145,-22,30]
ax.set_extent(extent)
ax.set_title('Nitrate Gradient Along Cruise TN413 in the Equatorial Pacific')
plt.savefig('TN413 NO3 Transect Gradient')
plt.show()
#Sources accessed: https://github.com/SciTools/cartopy/issues/922
```

```python
#Make a contour map to visualize nitrate over broad Equtorial Pacific area
#Read in Climatology Data as .csv, create filepath as df, and display
filepath='NO3 concentration (climatology) (3).csv'
picoplank_df = pd.read_csv(filepath, delimiter= ',')
print(picoplank_df.columns.values)
display(picoplank_df)
#Use tricontourf to make contour map using our 1D data
NO3_data= picoplank_df['NO3_darwin_clim']
lat = picoplank_df['lat']
lon = picoplank_df['lon']
fig = plt.figure(figsize=(10,7))
cntr = plt.tricontourf(lat,lon,NO3_data,cmap='plasma')
bar=fig.colorbar(cntr)
bar.set_label('Nitrate Concentration (mmol/L)')
#Add axes titles
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Nitrate Concentrations in the Equatorial Pacific')
plt.savefig('Nitrate Concentration Contour')
plt.show()

#Learned how to use https://matplotlib.org/stable/gallery/images_contours_and_fields/tricontour_demo.html & Week 5 pre lecture notes
```
